```python
companies['Date Joined'] = pd.to_datetime(companies['Date Joined'])  # Converting the 'Date Joined' column to datetime format
companies  # Displaying the updated DataFrame with the 'Date Joined' column converted to datetime
```

```python
companies.info()
```

```python
companies['Join_year'] = companies['Date Joined'].dt.year  
# Extracting the year from the 'Date Joined' column and creating a new column 'Join_year'
```

```python
companies.head()
```

```python
companies_sampled = companies.sample(n=50, random_state=40) 
# Sampling 50 random rows from the DataFrame with a random state of 40
```

```python
companies_sampled.head()
```

```python
companies_sampled['year_to_unicorn'] = companies_sampled['Join_year'] - companies_sampled['Year Founded']  
# Calculating the number of years it took for each sampled company to become a unicorn
grouped = companies_sampled[['Industry', 'year_to_unicorn']].groupby('Industry').max().sort_values('year_to_unicorn')  
# Grouping the data by industry and finding the maximum years to unicorn
grouped  
# Displaying the grouped data showing the maximum years to unicorn for each industry
```

```python
plt.bar(data=grouped, x=grouped.index, height=grouped['year_to_unicorn'])  
# Creating a bar plot with industry names on the x-axis and years to unicorn on the y-axis
plt.title('Bar plot of maximum years taken by company to become unicorn per industry (from sample)')  
# Setting the title of the plot
plt.xlabel('Industry')  # Labeling the x-axis
plt.ylabel('Years to Become a Unicorn')  # Labeling the y-axis
plt.xticks(rotation=45, horizontalalignment='right')  # Rotating x-axis labels for better readability
plt.show()  # Displaying the plot
```

```python
companies_sampled.head()
```

```python
companies_sampled['Valuation_int'] = companies_sampled['Valuation'].str[1:-1].astype(int)  
# Converting the 'Valuation' column to integer after removing the dollar sign and 'B' suffix
companies_sampled.info()  
# Displaying information about the DataFrame to confirm the changes
```

```python
grouped = companies_sampled[['Industry', 'Valuation_int']].groupby('Industry').sum('Valuation_int').sort_values('Valuation_int')  
# Grouping the data by industry and summing the valuations
grouped  # Displaying the grouped data showing the total valuation for each industry
```

```python
companies['Date Joined'] = pd.to_datetime(companies['Date Joined'])  
# Converting the 'Date Joined' column to datetime format
```

```python
companies['Month Joined'] = companies['Date Joined'].dt.month_name()  
# Extracting the month name from the 'Date Joined' column and creating a new column 'Month Joined'
```

```python
companies['Years To Join'] = companies['Join_year'] - companies['Year Founded']  
# Calculating the number of years it took for each company to join the unicorn club from its founding date
companies.head(10)
```

```python
companies['Date Joined'].dt.year.value_counts()  # Counting the frequency of unicorn company joins by year
```

```python
companies_mask = companies['Date Joined'].dt.year == 2021  
# Creating a mask to filter companies that joined in the year 2021
companies_2021 = companies[companies_mask].reset_index()  
# Creating a new DataFrame containing companies that joined in 2021 and resetting the index
```

```python
companies_2021.head(10)
```

```python
companies_2021['Week Joined'] = companies_2021['Date Joined'].dt.isocalendar().week  
# Extracting the week number from the 'Date Joined' column and creating a new column 'Week Joined'
```

```python
companies_2021.head(10)
```

```python
companies_2021_joined = companies_2021.groupby(by='Week Joined')['Company'].count().reset_index().rename(columns={"Company":"Company Count"})  
# Grouping the data by week joined and counting the number of companies in each week
companies_2021_joined.sort_values(by='Company Count')  
# Sorting the DataFrame by 'Company Count' in ascending order
```

```python
plt.figure(figsize=(20, 6))  # Setting the figure size for the plot
sns.barplot(data=companies_2021_joined, y='Company Count', x='Week Joined')  # Creating a bar plot
plt.title('Week_2021_joined')  # Adding a title to the plot
plt.xticks(rotation=45, horizontalalignment='left')  # Rotating x-axis labels for better readability
plt.yticks(range(0, 20, 1))  # Setting y-axis tick marks
plt.show()  # Displaying the plot
```
