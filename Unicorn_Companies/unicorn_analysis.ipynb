```python
companies_mask = companies['Date Joined'].dt.year == 2021  
# Creating a mask to filter companies that joined in the year 2021
companies_2021 = companies[companies_mask].reset_index()  
# Creating a new DataFrame containing companies that joined in 2021 and resetting the index
```

```python
companies_2021.head(10)
```

```python
companies_2021['Week Joined'] = companies_2021['Date Joined'].dt.isocalendar().week  
# Extracting the week number from the 'Date Joined' column and creating a new column 'Week Joined'
```

```python
companies_2021.head(10)
```

```python
companies_2021_joined = companies_2021.groupby(by='Week Joined')['Company'].count().reset_index().rename(columns={"Company":"Company Count"})  
# Grouping the data by week joined and counting the number of companies in each week
companies_2021_joined.sort_values(by='Company Count')  
# Sorting the DataFrame by 'Company Count' in ascending order
```

```python
plt.figure(figsize=(20, 6))  # Setting the figure size for the plot
sns.barplot(data=companies_2021_joined, y='Company Count', x='Week Joined')  # Creating a bar plot
plt.title('Week_2021_joined')  # Adding a title to the plot
plt.xticks(rotation=45, horizontalalignment='left')  # Rotating x-axis labels for better readability
plt.yticks(range(0, 20, 1))  # Setting y-axis tick marks
plt.show()  # Displaying the plot
```

```python
# Filter companies that joined in 2020
companies_2020 = companies[companies['Date Joined'].dt.year == 2020]

# Concatenate companies joined in 2020 with companies joined in 2021
companies_2020_2021 = pd.concat([companies_2020, companies_2021.drop(columns={'index','Week Joined'})])

# Extract the quarter joined from the 'Date Joined' column
companies_2020_2021["Quarter Joined"] = companies_2020_2021["Date Joined"].dt.to_period('Q').dt.strftime('%Y-Q%q')

# Convert valuation column to float
companies_2020_2021['Valuation'] = companies_2020_2021['Valuation'].str[1:-1].astype(float)

# Compute the mean valuation per quarter
companies_2020_2021_Quarter = companies_2020_2021.groupby(by='Quarter Joined')['Valuation'].mean().reset_index().rename(columns={'Valuation':'Valuation Per Quarter'})

# Reset index
companies_2020_2021_Quarter.index += 1

# Display the DataFrame
companies_2020_2021_Quarter
```

```python
# Extract the year and quarter components from the 'Quarter Joined' column
companies_2020_2021_Quarter['Year Joined'] = companies_2020_2021_Quarter['Quarter Joined'].str[:4]
companies_2020_2021_Quarter['Quarter Joined'] = companies_2020_2021_Quarter['Quarter Joined'].str[-2:]

# Display the modified DataFrame
companies_2020_2021_Quarter
```

```python
# Create a bar plot to compare the average valuation per quarter for companies that joined in 2020 vs 2021
plt.figure(figsize=(20,6))
sns.barplot(data=companies_2020_2021_Quarter, x="Quarter Joined", y='Valuation Per Quarter', hue='Year Joined')
plt.title('Company Valuation Per Quarter for 2020 vs 2021')
plt.show()
```

```python
# Extract the quarter joined from the 'Date Joined' column
companies_Quarter['Quarter Joined'] = companies_Quarter['Date Joined'].dt.to_period('Q').dt.strftime('%Y-Q%q')

# Convert valuation column to float
companies_Quarter['Valuation'] = companies_Quarter['Valuation'].str.strip('$B').astype(float)

# Display the modified DataFrame
companies_Quarter
```

```python
# Group by quarter joined and calculate the mean valuation per quarter
companies_Quarter = companies_Quarter.groupby(by='Quarter Joined')['Valuation'].mean().reset_index().rename(columns={'Valuation':'Valuation Per Quarter'})

# Display the DataFrame
companies_Quarter
```

```python
# Extract the year and quarter components from the 'Quarter Joined' column
companies_Quarter['Year Joined'] = companies_Quarter['Quarter Joined'].str[:4]
companies_Quarter['Quarter Joined'] = companies_Quarter['Quarter Joined'].str[-2:]

# Remove the first row (index 0) since it may contain incomplete data
companies_Quarter = companies_Quarter[1:]

# Reset index
companies_Quarter.reset_index(inplace=True, drop=True)
companies_Quarter.index += 1

# Display the modified DataFrame
companies_Quarter.head()
```

```python
companies_Quarter=companies_Quarter.sort_values(by='Year Joined')
```

```python
# Define the desired order of quarters
quarter_order = ['Q1', 'Q2', 'Q3', 'Q4']

# Create the bar plot with the specified order
plt.figure(figsize=(30, 15))
sns.barplot(data=companies_Quarter, x='Quarter Joined', y='Valuation Per Quarter', hue='Year Joined', order=quarter_order)
plt.legend(loc='upper right', fontsize=20)
plt.xlabel('Quarter', size=20)
plt.ylabel('Valuation Per Quarter', size=20)
plt.xticks(size=20)
plt.yticks(size=20)
plt.show()
```

```python
plt.figure(figsize=(15, 8))
sns.barplot(data=companies_Quarter, x='Quarter Joined', y='Valuation Per Quarter', hue='Year Joined', order=quarter_order)

# Add labels over each bar with respective year
for p, year in zip(plt.gca().patches, companies_Quarter['Year Joined']):
    height = p.get_height()
    plt.gca().text(p.get_x() + p.get_width() / 2, height + 0.5, f'{year}', ha='center', fontsize=10, rotation=90)

plt.legend(loc='upper right', fontsize=12)
plt.title('Average Valuation Per Quarter by Year', fontsize=16)
plt.xlabel('Quarter', fontsize=14)
plt.ylabel('Average Valuation Per Quarter', fontsize=14)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
plt.show()
## This was made through ChatGPT adding labels.
```
