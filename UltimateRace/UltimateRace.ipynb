```python
# Importing pandas library for data manipulation
import pandas as pd
# Importing seaborn library for data visualization
import seaborn as sns
# Importing numpy library for numerical computing
import numpy as np
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv('TWO_CENTURIES_OF_UM_RACES.csv')
```

```python
df.head(10)  # This will display the first ten rows of the DataFrame
```

```python
# Displaying concise summary information about the DataFrame
df.info()
```

```python
# Displaying the data types of each column in the DataFrame
df.dtypes
```

```python
# Counting the occurrences of unique values in the column "Event distance/length"
df.value_counts('Event distance/length')
```

```python
# Counting the occurrences of unique values in the "Year of event" column, sorted in descending order
df.value_counts('Year of event',sort=True).head(20)
```

```python
# Extracting the last four characters from the "Event name" column
df['Event name'].str[-4:-1]
```

```python
# Filtering rows:
# 1. By event distance ('50Km', '50mi', '50Mi', '50km')
# 2. For the year 2021
# 3. By event name ending with 'USA'
df[df['Event distance/length'].isin(['50Km','50mi','50Mi','50km'])
   &(df['Year of event']==2021)
   &(df['Event name'].str[-4:-1]=='USA')]
```

```python
df_50_Usa=df[df['Event distance/length'].isin(['50Km','50mi','50Mi','50km'])
   &(df['Year of event']==2021)
   &(df['Event name'].str[-4:-1]=='USA')]
```

```python
df_50_Usa
```

```python
df_50_Usa.shape
```

```python
# Removing the last six characters from the 'Event name' column
df_50_Usa['Event name']=df_50_Usa['Event name'].str[0:-6]
```

```python
df_50_Usa
```

```python
# Calculating the age of athletes by subtracting their year of birth from 2021
df_50_Usa['athlete_age'] = 2021 - df_50_Usa['Athlete year of birth']
```

```python
df_50_Usa
```

```python
# Removing the last two characters from the strings in the 'Athlete performance' column
df_50_Usa['Athlete performance'].str[0:-2]
```

```python
# Extracting the first word from each string in the 'Athlete performance' column
df_50_Usa['Athlete performance']=df_50_Usa['Athlete performance'].str.split(' ').str.get(0)
```

```python
df_50_Usa
```

```python
df_50_Usa.value_counts('Athlete country')
```

```python
# Creating a boolean mask to filter rows where the athlete's country is 'USA'
df_50_Usa_mask=df_50_Usa['Athlete country']=='USA'
```

```python
df_50_Usa=df_50_Usa[df_50_Usa_mask]
```

```python
df_50_Usa.value_counts('Athlete country')
```

```python
df_50_Usa.dtypes
```

```python
# Dropping columns 'Athlete club', 'Athlete country', 'Athlete year of birth', 'Athlete age category' from the DataFrame
df_50_Usa=df_50_Usa.drop(['Athlete club','Athlete country','Athlete year of birth','Athlete age category'],axis=1)
```

```python
df_50_Usa.head()
```

```python
# Calculating the number of missing values in each column of the DataFrame
df_50_Usa.isna().sum()
```

```python
# Filtering rows where the 'athlete_age' column contains missing values (NaNs)
df_50_Usa[df_50_Usa['athlete_age'].isna()==1]
```

```python
# Dropping rows containing any missing values (NaNs) from the DataFrame
df_50_Usa=df_50_Usa.dropna()
```

```python
df_50_Usa.isna().sum()
```

```python
# Filtering rows that are duplicates in the DataFrame
df_50_Usa[df_50_Usa.duplicated()==True]
```

```python
# Resetting the index of the DataFrame while dropping the existing index
df_50_Usa.reset_index(drop=True)
```

```python
df_50_Usa.dtypes
```

```python
# Converting the data type of the 'athlete_age' column to integer
df_50_Usa['athlete_age']=df_50_Usa['athlete_age'].astype(int)
# Converting the data type of the 'Athlete average speed' column to float
df_50_Usa['Athlete average speed']=df_50_Usa['Athlete average speed'].astype(float)
```

```python
df_50_Usa.dtypes
```

```python
df_50_Usa.head()
```

```python
# Renaming columns in the DataFrame
df_50_Usa = df_50_Usa.rename(columns={
    'Year of event': 'year',
    'Event dates': 'race_day',  
    'Event name': 'race_name',
    'Event distance/length': 'race_length',
    'Event number of finishers': 'number_of_finisher',
    'Athlete performance': 'athlete_performance',
    'Athlete gender': 'gender',
    'Athlete average speed': 'athlete_average_speed',
    'Athlete ID': 'athlete_ID',
})
```

```python
df_50_Usa
```

```python
df_50_Usa.dtypes
```

```python
# Selecting specific columns from df_50_Usa and creating a new DataFrame df_50_Usa_2
df_50_Usa_2 = df_50_Usa[['race_day', 'race_name', 'race_length', 'number_of_finisher',
                          'athlete_average_speed', 'athlete_ID', 'gender',
                          'athlete_performance', 'athlete_age']]
```

```python
df_50_Usa_2
```

```python
# Creating a histogram to visualize the distribution of race lengths
sns.histplot(df_50_Usa_2['race_length'])
```

```python
# Creating a histogram to visualize the distribution of race lengths, with gender differentiation
sns.histplot(data=df_50_Usa_2, x='race_length', hue='gender')
```

```python
df_50_Usa_2.value_counts('gender')
```

```python
# Creating a boolean mask to filter rows where the gender is either 'M' or 'F'
df_50_Usa_2_mask=(df_50_Usa_2['gender']=='M')  |(df_50_Usa_2['gender']=='F')
```

```python
df_50_Usa_2=df_50_Usa_2[df_50_Usa_2_mask]
```

```python
df_50_Usa_2.value_counts('gender')
```

```python
# Creating a histogram to visualize the distribution of race lengths with gender differentiation
sns.histplot(data=df_50_Usa_2, x='race_length', hue='gender')
```

```python
# Filtering the DataFrame for races with a race_length of 50 miles and
# Creating a distribution plot for athlete_average_speed
sns.displot(df_50_Usa_2[df_50_Usa_2['race_length']=='50mi']['athlete_average_speed'])
```

```python
# Creating a violin plot to visualize the distribution of athlete average speed across different race lengths, split by gender
sns.violinplot(data=df_50_Usa_2,x='race_length',y='athlete_average_speed',hue='gender',
               split=True,inner='quarts')
```

```python
# Creating a scatter plot with a linear regression line to visualize the relationship between athlete age and athlete average speed, colored by gender
sns.lmplot(data=df_50_Usa_2, x='athlete_age',y='athlete_average_speed',hue='gender')
```

```python
df_50_Usa_2.dtypes
```

```python
# Calculating the mean athlete average speed for each combination of race length and gender
df_50_Usa_2.groupby(['race_length','gender'])['athlete_average_speed'].mean()
```

```python
# Calculating the mean athlete age for each combination of race length and gender
df_50_Usa_2.groupby(['race_length','gender'])['athlete_age'].mean()
```
